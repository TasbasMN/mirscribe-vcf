```python
import pandas as pd

```

```python
latest = "latest.txt"

df = pd.read_csv(latest, sep='\t')
df.head()
```

```python
max_values = df.iloc[:, 2:].max(axis=1)
mask = max_values > 0.5

df.loc[mask, 'mutsig'] = df.iloc[:, 2:].idxmax(axis=1)[mask]

df = df.iloc[:, [0, 1, -1]]
df.fillna("SBS0", inplace=True)

```

```python
df.to_csv("probabilities_december.csv", index=False)
```
