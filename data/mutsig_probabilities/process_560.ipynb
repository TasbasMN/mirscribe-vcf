```python
import pandas as pd
```

```python
df = pd.read_csv("BreastCancer560_prob.txt", sep="\t")

df.head()
```

```python
df['mutsig'] = pd.Series(dtype=object)  # Initialize with object data type


# Assign mutsig values based on maximum probabilities
max_values = df.iloc[:, 2:].max(axis=1)
mask = max_values > 0.5

df.loc[mask, 'mutsig'] = df.iloc[:, 2:].idxmax(axis=1)[mask]

df = df.iloc[:, [0, 1, -1]]

df.fillna("SBS0", inplace=True)

df.head()
```

```python
df.to_csv("probabilities_560.csv", index=False)
```
